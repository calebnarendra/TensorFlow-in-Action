## 2.1 Multilayer Perceptron (MLP)
MLP adalah jenis neural network dasar yang terdiri dari:
* **Input Layer**: Menerima data masukan.
* **Hidden Layers**: Lapisan tersembunyi (satu atau lebih) di mana setiap neuron memiliki **bobot (weights)** dan **bias** untuk komputasi.
* **Output Layer**: Memberikan hasil akhir.

**Proses Transformasi:**
* Menggunakan **fungsi Sigmoid** (fungsi nonlinear sederhana) untuk transformasi data.
* Menggunakan **Softmax Activation** pada lapisan akhir untuk menormalisasi nilai linear menjadi probabilitas (total nilai = 1).
* Di TensorFlow 2, proses ini diimplementasikan menggunakan `tf.Variable`, fungsi aktivasi, dan perkalian matriks `tf.matmul`.

## 2.2 Model Eksekusi TensorFlow 2
Berbeda dengan versi pendahulunya, TensorFlow 2 menggunakan **Imperative Style Execution** (Eager Execution):
* **Deklarasi dan Eksekusi**: Terjadi secara bersamaan, sehingga lebih mudah untuk *debugging*.
* **AutoGraph (@tf.function)**: Menggunakan *decorator* `@tf.function` untuk mengubah fungsi Python biasa menjadi grafik data-flow TensorFlow secara otomatis. Ini memberikan performa tinggi tanpa kompleksitas pembuatan graf manual.

---

## 2.3 Building Blocks Utama
Ada tiga elemen inti dalam ekosistem TensorFlow:
1.  **tf.Variable**: Digunakan untuk mendefinisikan parameter model yang dapat berubah (seperti bobot). Harus memiliki dimensi (*shape*), nilai awal, dan jenis data (*dtype*).
2.  **tf.Tensor**: Output dari operasi TensorFlow. Tensor adalah struktur data multidimensi (skalar, vektor, matriks, hingga dimensi tak terbatas).
3.  **tf.Operation**: Melaksanakan tugas aritmetik. Sering diiringi teknik **Broadcasting** (menyamakan dimensi secara otomatis) dan **keepdims** untuk menjaga konsistensi dimensi tensor.

## 2.4 Operasi Matriks & Manipulasi Dimensi
Data kompleks seringkali perlu disederhanakan sebelum diproses:
* **Penyederhanaan**: Contohnya mengubah gambar RGB menjadi Grayscale sebelum dilakukan `tf.matmul`.
* **tf.squeeze**: Digunakan untuk membuang dimensi yang berukuran 1 (dimensi yang tidak diperlukan) agar struktur tensor lebih ringkas.

---

## 2.5 Operasi Konvolusi (Convolution)
Konvolusi adalah inti dari CNN untuk mendeteksi fitur seperti tepi (*edges*) pada gambar.
* **Mekanisme**: Menggeser **Kernel** (filter) di atas data input untuk menghasilkan *feature map* baru.
* **Filter Laplacian**: Contoh matriks $3 \times 3$ yang digunakan untuk mendeteksi tepi dengan menghasilkan nilai nol pada area yang seragam.
* **Implementasi**: Menggunakan `tf.nn.convolution` yang memerlukan format input dan filter dengan dimensi yang spesifik.



---

## 2.6 Operasi Pooling
Setelah deteksi tepi, pooling digunakan untuk mereduksi ukuran data:
* **Tujuan**: Mengurangi tinggi dan lebar gambar (misal: dibagi dua) untuk mempercepat komputasi dan mengurangi parameter.
* **Cara Kerja**: Mirip konvolusi tetapi **tanpa bobot kernel**. Pooling langsung mengambil nilai (maksimum atau rata-rata) dari area yang tumpang tindih.
* Operasi ini tidak menggunakan filter pemroses, melainkan murni agregasi data spasial.
